```python
# Import libraries
import matplotlib.pyplot as plt

# Data
streams = [1, 3]
H_values = [230, 240]  # Values of |H| [MW]
T1_values = [20, 135]  # Supply and target temperatures for Stream 1
T3_values = [80, 140]  # Supply and target temperatures for Stream 3

# Create the figure and axes
plt.figure(figsize=(8,6))

# Plot the first series (Stream 1)
plt.plot(H_values, T1_values, label="Cold Stream 1", marker='o', color='blue', linestyle='-', linewidth=2)

# Plot the second series (Stream 3)
plt.plot(H_values, T3_values, label="Cold Stream 3", marker='o', color='red', linestyle='-', linewidth=2)

# Add horizontal lines for each temperature
plt.axhline(y=20, color='blue', linestyle='--', linewidth=1, label="20C (Stream 1 Ts)")
plt.axhline(y=135, color='blue', linestyle='--', linewidth=1, label="135C (Stream 1 Tm)")
plt.axhline(y=80, color='red', linestyle='--', linewidth=1, label="80C (Stream 3 Ts)")
plt.axhline(y=140, color='red', linestyle='--', linewidth=1, label="140C (Stream 3 Tm)")

# Add labels and title
plt.title("Process Streams: Temperatures vs. H", fontsize=14)
plt.xlabel("|H| [MW]", fontsize=12)
plt.ylabel("Temperature [C]", fontsize=12)

# Show legend
plt.legend()

# Show grid
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for intervals as streams
H_values1 = [120, 330, 20]  # |H| [MW] values for the first curve
T1_values1 = [20, 80]       # Temperatures for Interval 1 of the first curve
T2_values1 = [80, 135]      # Temperatures for Interval 2 of the first curve
T3_values1 = [135, 140]     # Temperatures for Interval 3 of the first curve

H_values2 = [45, 405, 60]   # |H| [MW] values for the second curve
T1_values2 = [30, 60]       # Temperatures for Interval 1 of the second curve
T2_values2 = [60, 150]      # Temperatures for Interval 2 of the second curve
T3_values2 = [150, 170]     # Temperatures for Interval 3 of the second curve

# Calculate x-axis values for each interval
def calculate_intervals(H_values, start_shift=0):
    x_start = [start_shift]
    x_end = [x_start[0] + H_values[0]]
    for i in range(1, len(H_values)):
        x_start.append(x_end[-1])
        x_end.append(x_start[-1] + H_values[i])
    return x_start, x_end

# Interpolation function
def get_interpolated_temperatures(x_range, x_start, x_end, T_values):
    y = np.zeros_like(x_range)
    for i in range(len(x_start)):
        mask = (x_range >= x_start[i]) & (x_range <= x_end[i])
        if len(T_values) > i + 1:
            y[mask] = np.interp(x_range[mask], [x_start[i], x_end[i]], [T_values[i], T_values[i + 1]])
    return y

# Find the shift needed for the temperature difference to be 10C
def find_shift_for_temp_difference(H_values1, T_values1, H_values2, T_values2, target_temp_diff, initial_shift=60):
    start_shift1 = initial_shift
    step = 1
    max_iterations = 1000

    for _ in range(max_iterations):
        x_start1, x_end1 = calculate_intervals(H_values1, start_shift=start_shift1)
        x_start2, x_end2 = calculate_intervals(H_values2)

        # Create continuous x ranges for interpolation
        x_range2 = np.linspace(min(x_start2), max(x_end2), 1000)
        y2 = get_interpolated_temperatures(x_range2, x_start2, x_end2, T_values2)
        y1_shifted = get_interpolated_temperatures(x_range2, x_start1, x_end1, T_values1)

        differences = np.abs(y2 - y1_shifted)
        delta_H_index = np.where(np.abs(differences - target_temp_diff) < 0.1)[0]

        # Check that Curve 1 is never greater than or equal to Curve 2
        if len(delta_H_index) > 0:
            if np.all(y1_shifted < y2):  # Ensure Curve 1 is less than Curve 2 at all points
                delta_H = x_range2[delta_H_index[0]]
                return start_shift1, delta_H

        start_shift1 += step

    return None, None

# Calculate the shift needed for the temperature difference to be 10C
target_temp_diff = 10
start_shift1, delta_H = find_shift_for_temp_difference(H_values1, T1_values1 + T2_values1[1:] + [T3_values1[-1]],
                                                      H_values2, T1_values2 + T2_values2[1:] + [T3_values2[-1]],
                                                      target_temp_diff)

if start_shift1 is not None:

    print(f"H where the temperature difference is 10C: {delta_H:.2f} MW")
else:
    print("Could not find the necessary shift")

# Create the figure and axes
plt.figure(figsize=(12, 6))

# Plot the first curve with the calculated shift
x_start1, x_end1 = calculate_intervals(H_values1, start_shift=start_shift1)
plt.plot([x_start1[0], x_end1[0]], [T1_values1[0], T1_values1[1]], label="Interval 1 - Cold fluid", marker='o', color='blue', linestyle='-', linewidth=2)
plt.plot([x_start1[1], x_end1[1]], [T2_values1[0], T2_values1[1]], label="Interval 2 - Cold fluid", marker='o', color='red', linestyle='-', linewidth=2)
plt.plot([x_start1[2], x_end1[2]], [T3_values1[0], T3_values1[1]], label="Interval 3 - Cold fluid", marker='o', color='green', linestyle='-', linewidth=2)

# Plot the second curve
plt.plot([x_start2[0], x_end2[0]], [T1_values2[0], T1_values2[1]], label="Interval 1 - Hot fluid", marker='o', color='cyan', linestyle='--', linewidth=2)
plt.plot([x_start2[1], x_end2[1]], [T2_values2[0], T2_values2[1]], label="Interval 2 - Hot fluid", marker='o', color='magenta', linestyle='--', linewidth=2)
plt.plot([x_start2[2], x_end2[2]], [T3_values2[0], T3_values2[1]], label="Interval 3 - Hot fluid", marker='o', color='orange', linestyle='--', linewidth=2)

# Plot the vertical line at the H where the difference is 10C
if delta_H is not None:
    plt.axvline(x=delta_H, color='black', linestyle='--', label='Difference = 10C')

# Add labels and title
plt.title("Combined Curves for T=10 C: Temperatures vs. H", fontsize=14)
plt.xlabel("|H| [MW]", fontsize=12)
plt.ylabel("Temperature [C]", fontsize=12)

# Show legend
plt.legend()

# Show grid
plt.grid(True)

# Show plot
plt.show()
```
