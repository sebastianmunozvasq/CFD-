### Plotting Velocity Profiles Using Python







```python
# Import the packages we need

# Numerical computation, vector and matrix generation, and various mathematical operations
import numpy as np

# Plotting
import matplotlib.pyplot as plt
```

## Prepare Computational Variables


```python
# Define the variables

# Typical pressure in a residential water pipe
P_0 = 102000  # Pa

# Discharge pressure to the atmosphere
P_L = 101325  # Pa

# Length of the pipe
L = 10  # m

# Viscosity of water
mu = 1e-3  # Pa * s

# Radius of the pipe
R = 1 * 0.0254  # m
```

First, we will evaluate the velocity at a specific position.


```python
# Example position
r = R / 2

# Compute the velocity
v_z = (P_0 - P_L) / (4 * mu * L) * R**2 * (1 - (r / R)**2)

# Print the result
print("v_z = %.3f m/s " % v_z)
```

```python
# Evaluate the velocity profile at all positions contained in the np.array r

v_z = (P_0 - P_L) / (4 * mu * L) * R**2 * (1 - (r / R)**2)

# Print the calculated velocities
print(v_z)
```

## Plotting


```python
# Basic syntax using plt.plot
plt.plot(r, v_z)
```

```python
# Initial plot
plt.plot(r, v_z)

# Add axis labels and specify their size
plt.xlabel('$r$ / m', size=14)
plt.ylabel('$v_z$ / m$s^{-1}$', size=14)

# Show the plot
plt.show()
```

## Improve Code and Plot Quality


```python
# Define the function to calculate velocity within the pipe
def vz_tuberia(r, R, P_0, P_L, mu, L):
    v_z = (P_0 - P_L) / (4 * mu * L) * R**2 * (1 - (r / R)**2)
    return v_z
```

```python
# Test the implementation
vz_tuberia(r, R, P_0, P_L, mu, L)
```

### Plotting the Profile for Three Different Pipe Radii


```python
# Define three radii and assign labels
R_list = [R / 2, R, 2 * R]
labels = ["$R/2$", "R", "2R"]
```

```python
# Iterate from i=0 to i=2
for i in range(len(R_list)):

    # Redefine the variable r
    r = np.linspace(0, R_list[i], 100)

    # Plot the velocity profile
    plt.plot(r, vz_tuberia(r, R_list[i], P_0, P_L, mu, L), label=labels[i])
    plt.legend(fontsize=12)

    # Annotate the pipe wall
    plt.axvline(R_list[i], ls='--', color='k')
    plt.annotate('R = %.3f m' % R_list[i], xy=(R_list[i] - 0.003, 15), size=14, rotation=90)

# Add axis labels and title with specified size
plt.title('Velocity Profile in a Circular Pipe', size=14)
plt.xlabel('$r$', size=14)
plt.ylabel('$v_z$', size=14)

# Show the plot
plt.show()
```
